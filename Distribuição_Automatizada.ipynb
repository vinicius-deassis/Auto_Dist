```python
import pandas as pd
#Setting whole lists
lista_envio = [
    'Ana Caroline Cabral - Brazil','Antonio Leal - Brazil', 'Arizza Goes - Brazil',
    'Carolayne dos Santos - Brazil','Dayane Santos - Brazil',
    'Eduardo Santos - Brazil','Gabrielly da Silva - Brazil', 'Guilherme Gomes - Brazil',
    'Heveline Lira - Brazil','Josue Leite - Brazil','Lucilenne Silva - Brazil',
    'Mariana Cavalcante - Brazil','Marianne da Silva - Brazil','Othon Gibson - Brazil',
    'Tacilla Magalhaes - Brazil', 'Diego Henrique - Brazil',

]
lista_planilhas = [
    'Ana Caroline Finalização 86','Antônio Finalização 86','Arizza Finalização 86',
    'Carolayne Finalização 86','Dayane Edite Finalização 86',
    'Eduardo Finalização 86','Gabrielly Finalização 86','Guilherme Finalização 86',
    'Heveline Finalização 86','Josue Jr Finalização 86','Lucilenne Finalização 86',
    'Mariana Finalização 86','Marianne Finalização 86','Othon Finalização 86',
    'Tacilla Finalização 86', 'Diego Finalização 86'
]
lista_moto = [
    'Camilla Mascarenhas - Brazil','Helder Junior - Brazil',
    'Kaio Rodrigues - Brazil', 'Manuela Araújo - Brazil', 'Michel Costa - Brazil',
    'Ana Carla Marinho - Brazil', 'Marcelle Teixeira - Brazil','Rayanne Vasconcelos - Brazil'
]
envio = [
    'Recrutamento/Ativação/Reativação - Camilla Mascarenhas',
    'Recrutamento/Ativação/Reativação - Helder Santos','Recrutamento/Ativação/Reativação - Kaio Barbosa',
    'Recrutamento/Ativação/Reativação - Manuela Araújo','Recrutamento/Ativação/Reativação - Michel Costa',
    'Recrutamento/Ativação/Reativação - Ana Carla Marinho','Recrutamento/Ativação/Reativação - Marcelle Teixeira',
    'Recrutamento/Ativação/Reativação - Rayanne Vasconcelos'
]
#Updating the people who are not going to receive leads
'''lista_envio.remove('Othon Gibson - Brazil')
lista_envio.remove('Guilherme Gomes - Brazil')
lista_planilhas.remove('Othon Finalização 86')
lista_planilhas.remove('Guilherme Finalização 86')
'''
lista_moto.remove('Camilla Mascarenhas - Brazil')
envio.remove('Recrutamento/Ativação/Reativação - Camilla Mascarenhas')

#functions
def send(planilha,df,vertical):
  #opens the spreadsheet
  sheet = gc.open(planilha)
  if vertical == 1:
    #sets the manipulated worksheet
    wsheet = sheet.worksheet('Sheet1')
    #receives how many columns are fulfilled
    length = len(wsheet.col_values(1))
    #pastes the dataframe in the first blank spot
    wsheet.update(f'A{length+1}',df.to_numpy().tolist())
  else:
    #same as the first but with the Moto pattern
    wsheet = sheet.worksheet('Recrutamento')
    length = wsheet.col_values(1)
    length = len(length)
    wsheet.update(f'A{length+1}',df.to_numpy().tolist())

def distribuir(vertical):
  sheet = gc.open(sheet_name)
  temp_list = []
  if vertical == 1:
    # 1 meaning the car vertical
    wsheet = sheet.worksheet('Carro')
    leads_length = wsheet.col_values(1)
    leads_length = len(leads_length)
    print(leads_length)
    rec = len(lista_envio)
    #sets how many leads for each recruiter, equally
    leads_per_rec = leads_length // rec
    total_send = int(rec * leads_per_rec)
    y = leads_length - total_send
    print(y)
    for i in range(0,leads_per_rec):
      for j in range(0,rec):
        temp_list.append(lista_envio[j])
    for i in range(0,y-1):
        temp_list.append(lista_envio[i])
    df = {'distribuition':temp_list}
    envio_df = pd.DataFrame(df)
    wsheet.update('F1','Dist')
    wsheet.update('F2',envio_df.to_numpy().tolist())
    print(leads_per_rec,'leads para cada')
  else:
    wsheet = sheet.worksheet('Moto')
    leads_length = wsheet.col_values(1)
    leads_length = len(leads_length)
    rec = len(lista_moto)
    leads_per_rec = leads_length // rec
    total_send = int(rec * leads_per_rec)
    y = leads_length - total_send
    print(y)
    for i in range(0,leads_per_rec):
      for j in range(0,rec):
        temp_list.append(lista_moto[j])
    for i in range(0,y-1):
        temp_list.append(lista_moto[i])
    df = {'distribuition':temp_list}
    envio_df = pd.DataFrame(df)
    wsheet.update('F1','Dist')
    wsheet.update('F2',envio_df.to_numpy().tolist())
    print(leads_per_rec,'leads para cada')

def filtrar(recrutador,planilha,vertical):
  sheet = gc.open(planilha)
  if vertical == 1:
    wsheet = sheet.worksheet('Carro')
    base = pd.DataFrame(wsheet.get_all_records())
    df = base.loc[base['Dist'] == recrutador]

    return df
  else:
    wsheet = sheet.worksheet('Moto')
    base = pd.DataFrame(wsheet.get_all_records())
    df = base.loc[base['Dist'] == recrutador]
    return df

def restante():
  df = filtrar("",sheet_name,1)
  send('database finalização carro',df,1)

print('Qual base você quer enviar \n1)Verificar \n2)Carro \n3)Moto \n4)Ambos' )
print('5Distribuir carro \n6)Distribuir carro \n7)Distribuir ambos\n')
choice = int(input())
if choice == 1:
  if len(lista_envio) == len(lista_planilhas):
    print('Tamanho da lista de carro ok')
  else:
    print('Erro na lista de carro')

  if len(envio) == len(lista_moto):
    print('Tamanho da lista de moto ok')
  else:
    print('Erro na lista de moto')

  try:
    for i in range(0,len(lista_planilhas)):
        teste = {'teste':[""]}
        df = pd.DataFrame(teste)
        send(lista_planilhas[i],df,1)
    print('Limites de linhas carro ok')
  except:
    print(f'Limite de linhas de {lista_planilhas[i]} é menor que a necessária')

  try:
    for i in range(0,len(envio)):
        teste = {'teste':[""]}
        df = pd.DataFrame(teste)
        send(envio[i],df,2)
    print('Limites de linhas de moto ok')
  except:
    print(f'Limite de linhas de {envio[i]} é menor que a necessária')
elif choice == 2:
  #distribuir(1)
  for i in range(0,len(lista_planilhas)):
    df = filtrar(lista_envio[i],sheet_name,1)
    df = df.drop('Dist', axis=1)
    send(lista_planilhas[i],df,1)
    print(lista_planilhas[i])

elif choice == 3:
  #distribuir(2)
  for i in range(0,len(envio)):
    df = filtrar(lista_moto[i],sheet_name,2)
    df = df.drop('Dist', axis=1)
    send(envio[i],df,2)
    print(envio[i])

elif choice == 4:
  #distribuir(1)
  for i in range(0,len(lista_planilhas)):
    df = filtrar(lista_envio[i],sheet_name,1)
    df = df.drop('Dist', axis=1)
    send(lista_planilhas[i],df,1)
    print(lista_planilhas[i])
  #distribuir(2)
  for i in range(0,len(envio)):
    df = filtrar(lista_moto[i],sheet_name,2)
    df = df.drop('Dist', axis=1)
    send(envio[i],df,2)
    print(envio[i])

elif choice == 5:
  distribuir(1)

elif choice == 6:
  distribuir(2)

elif choice == 7:
  distribuir(1)
  distribuir(2)
```
